## Dekoratorji

Dekoratorji se uporabljajo za dopolnjevanje uporabnosti druge funkcije, brez speminjanja druge funkcije.


```python
def moj_dekorator(func):                          #  func = funkcija, ki jo dekoriramo
    def ovojnica(*args,**kwargs):                 # ovojnico (wrapper) potrebujemo,da se nova funkcija ne sproži ko kličemo dekorator
        print("Pred klicem")                      # *args,**kwargs potrebujemo, da delajo argumenti v originalni funkciji
        func(*args,**kwargs)                      #  pokličemo originalno funkcijo
        print("Po klicu")
    return ovojnica

@moj_dekorator
def pozdravi(ime):
    print(f"Pozdravljen {ime} ")

pozdravi("Andrej")
```

### Dedovanje (inheritance)

Dedovanje omogoča prenos atributov in metod iz enega razreda (starš) na drugega (otrok).

Če želimo spremeniti dedovano metodo, jo samo ponovno definiramo v izbranem razredu.

```python
class Žival:

    def __init__(self,ime):
        self.ime=ime
        self.je_živa=True

    def jesti(self):
        print(f"{self.ime} je hrano.")

    def spanje(self):
        print(f"{self.ime} spi.")


class Pes(Žival):  # Razred Pes deduje od razreda Žival 

    def lajanje(self):
        print(f"{self.ime} je zalajal")

class Mačka(Žival):
    pass

class Miš(Žival):
    pass

pes= Pes("Capi")
mačka=Mačka("Predl")
miš=Miš("Sapra")


pes.jesti()
pes.lajanje()
```

### Dedovanje od več razredov in večnivojsko dedovanje

Dedovanje od več razredov je ko razred deduje od več različnih razredov. `A(B,C)`

Pri več nivojskem dedovanju pa starš deduje od starša.  


```python
class Žival:

    def __init__(self,ime):
        self.ime=ime
        
    def jesti(self):
        print(f"Žival je hrano.")

    def spati(self):
        print(f"Žival spi.")


class Hrana(Žival):
    def pobeg(self):
        print("Ta žival je pobegnila")

class Predator(Žival):
    def lov(self):
        print("Ta žival lovi")

class Zajec(Hrana):    #deduje od Hrane in Živali
    def skače(self):
        print(f"{self.ime} skače.")

class Jastreb(Predator):  #deduje od Predatorja in Živali
    pass

class Riba(Hrana, Predator):   #Riba deduje tako od Hrane kot od Predatorja in Živali
    pass


zajec=Zajec("zajko")
jastreb=Jastreb("miha")
riba=Riba("hadock")



zajec.skače()
```

### Abstraktni razredi

To so razredi, ki nimajo instanc. Služijo samo kot starši.
Vsebujejo abstraktne metode.
- ne morejo narediti instance.
- nujno potrebujejo otroka za dedovanje abstraktnih metod.
- abstraktne metode morajo biti nujno podedovane, če ne Error.
- abstraktne metode morajo biti dokončno definirane šele v otroku.
- eden izmed namenov je, da se pri dedovanju ne pozabi na pomembne metode, ki bi jih razred potreboval.

```python
from abc import ABC, abstractmethod # ABC=abstract base classes


class Vozilo(ABC):
    
    @abstractmethod     #abstraktne metode definiramo z dekoratorjem
    def pelji(self):
        pass

    @abstractmethod
    def stop(self):
        pass


#vozilo=Vozilo()

#----->TypeError: Can't instantiate abstract class Vozilo without an implementation for abstract methods 'pelji', 'stop'

class Avto(Vozilo):    #Ker deduje od Vozilo mora nujno vsebovatei vse abstraktne metode
    def pelji(self):
        print("Avto se pelje")

    def stop(self):
        print("Avto stoji")

avto=Avto()

avto.pelji()
avto.stop()
```

### Super()

Z funkcijo `super()` se doda oz. spremeni podedovano metodo.


```python

class Oblika:
    def __init__(self,barva,zapolnjen):
        self.barva=barva
        self.zapolnjen=zapolnjen
        
    def opis(self):
        print(f"Oblika je {self.barva} in {"zapolnjen" if self.zapolnjen else "ni zapolnjen." }")


class Krog(Oblika):
    def __init__(self,barva,zapolnjen, polmer):
        super().__init__(barva,zapolnjen)            #Kopira vsebino starševskega __init__
        self.polmer=polmer

    def opis(self):
        super().opis()                            #Kopira vsebino metode opis v nov opis, da se mu lako še doda stvari.
        print(f"Obseg je {3.14 * self.polmer}")

class Trikotnik(Oblika):
    def __init__(self,barva,zapolnjen, širina, višina):
        super().__init__(barva,zapolnjen) 
        self.širina=širina
        self.višina=višina

class Kvadrat(Oblika):
    def __init__(self,barva,zapolnjen, širina):
        super().__init__(barva,zapolnjen) 
        self.širina=širina


krog=Krog("Rdeč", True, 45)
trikotnik=Trikotnik("Moder", False, 45, 50)

trikotnik.opis()
krog.opis()
```

### Polimorfizem

- objekt ima lahko več form.

Torej v tem primeru je pizza krog. Krog pa je oblika. Zato je pizza tudi oblika.

```python
from abc import ABC, abstractmethod

class Oblika:
    

    @abstractmethod
    def površina(self):
        pass


class Krog(Oblika):
    def __init__(self, polmer):
        self.polmer=polmer

    def površina(self):
        return 3.14*self.polmer**2
        

class Trikotnik(Oblika):
    def __init__(self, stranica, višina):

        self.stranica=stranica
        self.višina=višina

    def površina(self):
        return (self.stranica*self.višina)/2

class Kvadrat(Oblika):
    def __init__(self, stranica):
        self.stranica=stranica

    def površina(self):
        return self.stranica**2


class Pizza(Krog):      #Pizza je krog. In preko kroga oblika.
    def __init__(self, nadev,polmer):
        self.nadev=nadev
        super().__init__(polmer)
        

oblike=[Krog(20),Trikotnik(13,14),Kvadrat(23), Pizza("sir", 10)]

for i in oblike:
    print(f"{i.površina()} kvadratnih centimetrov")   
```

### Gnezdeni razredi 
- grupiranje razredov, ki so si podobni
- zmanjša možnosti za konflikte z imenovanjem

```python
class Podjetje:

    class Zaposleni:
        def __init__(self,ime,pozicija):
            self.ime=ime
            self.pozicija=pozicija

        def podatki(self):
            return f"{self.ime} dela {self.pozicija}."


    def __init__(self, ime):
        self.ime=ime
        self.zaposleni=[]

    def dodaj_zaposlenega(self, ime, pozicija):
        nov_zaposleni=self.Zaposleni(ime, pozicija)
        self.zaposleni.append(nov_zaposleni)

    def naštej_zaposlene(self):
        for i in self.zaposleni:
            print(i.podatki())


podjetje=Podjetje("Umbrela Armory")
podjetje.dodaj_zaposlenega("Leon Kenedy", "Pištole")
podjetje.dodaj_zaposlenega("Andrej", "Fotograf")

podjetje.naštej_zaposlene()

```

### Statične metode

To so metode, ki pripadajo razredu ne pa instanci razreda (objektu).

Uporabljajo se, ko metoda ne potrebuje dostopa do razrednih podatkov.

Uporablja se kot splošna pomožna metoda.



```python
class Zaposleni:
    
    def __init__(self, ime, pozicija):
        self.ime=ime
        self.pozicija=pozicija

    def dobi_podatke(self):
        return f"{self.ime} dela {self.pozicija}"     # Klasična razredna metoda
    
    @staticmethod                                                 # Ta dekorator potrebujemo, če hočemo da metoda deluje tudi z instancami
    def obstaja(pozicija):                                         # Statična metoda
        obstajajoče_pozicije=["Meneger", "Kuhar", "Hišnik"]    

        if pozicija in obstajajoče_pozicije:
            return True
        else:
            return False


print(Zaposleni.obstaja("Meneger"))
```

### Razredne metode

Izvijajajo operacije, ki so povezane z razredom samim.

Imajo `cls` kot prvi parameter, ki prdstavlja celoten razred.

```python
class Študent:

    število=0


    def __init__(self, ime,ocena):
        self.ime=ime
        self.ocena=ocena
        Študent.število+=1
        
    def dobi_podatke(self):                      # Klasična razredna metoda
        return f"{self.ime} ima {self.ocena}"
    

    @classmethod
    def dobi_število(cls):                               # razredna metoda
        return f"Število študentov: {cls.število}"
    

študent1=Študent("Leon", 10)

print(študent1.dobi_podatke())

print(Študent.dobi_število())
```

### Vgrajene metode (Magic methods/Dunder methods)

Prepoznamo jih po __ __

Uporabljajo se za prilagajanja in definiranja obnašanja objektov.

```python
class Knjiga:

    def __init__(self, naslov, avtor,št_strani): 
        self.naslov=naslov
        self.avtor=avtor
        self.št_strani=št_strani

    def __str__(self):                   # Brez te metode print() poda pomnilniški naslov<__main__.Knjiga object at 0x00000227EF141400>
        return f"{self.naslov} od {self.avtor}"
    
    def __eq__(self, drugi):                                                  # Brez te metode bi bila primerjava vedno False, pa tudi če bi dva objekta imela vse atribute enake
        return self.naslov == drugi.naslov and self.avtor == drugi.avtor      # Izberemo kateri atributi so relevanti za pšrimerjavo dveh objektov.

    def __lt__(self, drugi):                                           # Za določitev atributv za manjše (<)
        return self.št_strani < drugi.št_strani

    def __gt__(self, drugi):                                           # Za določitev atributv za večje (>)
        return self.št_strani > drugi.št_strani
    
    def __add__(self, drugi):                                          # Za določitev atributv za seštevanje (+)
        return self.št_strani + drugi.št_strani
    
    def __contains__(self, keyword):                                   # Za določitev atributa in besede ali vrednosti  za pregled vsebovanja
        return keyword in self.naslov
    
    def __getitem__(self, key):                                        # Za iskanje atributov z indeksiranjem   
        if key == "naslov":
            return self.naslov
        elif key == "avtor":
            return self.avtor
        elif key == "št_strani":
            return self.št_strani
        else:
            return f"Ključ `{key}` ni bil najden."
    



knjiga1=Knjiga("Bratovščina Prstana", "Tolkien", 450)
knjiga2=Knjiga("Stolpa", "Tolkien", 470)
knjiga3=Knjiga("Kraljeva vrnitev", "Tolkien", 500)
knjiga4=Knjiga("Bratovščina Prstana", "Tolkien", 450)
        

print(knjiga1)  
print(knjiga1 == knjiga4)
print(knjiga1<knjiga2)
print(knjiga1>knjiga2)
print(knjiga1+knjiga2)
print("Stolpa" in knjiga2)
print(knjiga2["naslov"])
print(knjiga1["velikost"])
```

### @property

Za definiranje metod kot lastnina (lahko se ji dostopa kot atribute).

Za dodajanje dodatnih stvari atributom.

Pomembno zelo za Error.

- getter za branje atributov
- setter za spreminjanje atributov
- deleter za brisanje atributov


```python
class Pravokotnik:
    def __init__(self, širina, višina):
        self._širina=širina                  # _ pomeni, da sta atributa zaščitena. To pa pomeni, da jih je priporočlivo dostopati samo čez @property metodi
        self._višina=višina

    @property                               # Za dodajanje atributom
    def širina(self):                       # Tu se širina navezuje na atribut self._širina
        return f"{self._širina:.1f} cm"

    @property                              # Za dodajanje atributom
    def višina(self):                      # Tu se širina navezuje na atribut self._višina
        return f"{self._višina:.1f} cm"
    
    @širina.setter                          # .setter se uporabla za spreminjanje atributov.
    def širina(self, nova_širina):
        if nova_širina > 0:
            self._širina=nova_širina
        else:
            print("Nova širina mora biti  večja od 0")

    @višina.deleter                         # .deleter se uporablja za brisanje atributov
    def višina(self):
        del self._višina
        print("Višina je bila zbrisana")
        print(self.višina)                # AttributeError: 'Pravokotnik' object has no attribute '_višina', ker je bila višina zbrisana

pravoktnik=Pravokotnik(3, 5)

pravoktnik.širina=0

print(pravoktnik.višina)
print(pravoktnik.širina)

del pravoktnik.višina   # za klicanje deleter metode
```
